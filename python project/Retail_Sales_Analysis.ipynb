```python
import pandas as pd
import numpy as np
import sqlite3
from pathlib import Path

try:
    from google.colab import files
except Exception:
    files = None

pd.set_option('display.float_format', lambda x: f'{x:,.2f}')
print("Pandas:", pd.__version__)
import sqlite3, sys
print("SQLite (via Python):", sqlite3.sqlite_version)
print("Python:", sys.version.split()[0])
```

```python
if files:
    uploaded = files.upload()
    filename = next(iter(uploaded))
else:

    filename = "retail_sales.csv.xlsx"

print("Using file:", filename)
```

```python
path = Path(filename)
ext = path.suffix.lower()

if ext in [".xlsx", ".xls"]:
    df_raw = pd.read_excel(path)
elif ext == ".csv":

    df_raw = pd.read_csv(path, encoding="utf-8")
else:
    raise ValueError("Unsupported file type. Please upload a .csv or .xlsx/.xls file.")

print("Shape:", df_raw.shape)
df_raw.head(3)
```

```python
print(df_raw.columns.tolist())
df_raw.info()
```

```python
def clean_col(col):
    col = str(col).strip().lower()
    col = col.replace(" ", "_").replace("-", "_")
    col = col.replace("%", "pct").replace("(", "").replace(")", "")
    return col

df = df_raw.copy()
df.columns = [clean_col(c) for c in df.columns]

col_map = {
    "order_date": "orderdate",
    "order_dt": "orderdate",
    "discount_pct": "discount_pct",
    "discount": "discount_pct",
}

df = df.rename(columns={c: col_map[c] for c in df.columns if c in col_map})

print(df.columns.tolist())
df.head(3)
```

```python
date_candidates = [c for c in df.columns if "order" in c and "date" in c]
if date_candidates:
    date_col = date_candidates[0]
else:

    if "orderdate" in df.columns:
        date_col = "orderdate"
    else:
        raise KeyError("Could not find an order date column. Please rename the date column to 'OrderDate'.")

df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
for c in ["quantity", "unitprice", "discount_pct"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

if "discount_pct" not in df.columns:
    df["discount_pct"] = 0.0

df["quantity"] = df["quantity"].fillna(0)
df["unitprice"] = df["unitprice"].fillna(0.0)
df["discount_pct"] = df["discount_pct"].fillna(0.0)

df["sales"] = df["quantity"] * df["unitprice"] * (1 - (df["discount_pct"]/100.0))

for c in ["orderid","customerid","productid","storeid"]:
    if c in df.columns:
        df[c] = df[c].astype(str)

df.head(5)
```

```python
print("Nulls by column:")
print(df.isnull().sum())

if date_col in df.columns:
    print("\nDate range:", df[date_col].min(), "", df[date_col].max())

for c in ["quantity","unitprice","discount_pct","sales"]:
    if c in df.columns:
        print(f"{c}: min={df[c].min():,.2f}, max={df[c].max():,.2f}, mean={df[c].mean():,.2f}")
```

```python
db_path = "retail.db"
conn = sqlite3.connect(db_path)

df.to_sql("orders", conn, if_exists="replace", index=False)

with conn:
    for idx_col in ["orderid", "customerid", "productid", "storeid"]:
        if idx_col in df.columns:
            conn.execute(f"CREATE INDEX IF NOT EXISTS idx_orders_{idx_col} ON orders({idx_col});")
    if date_col in df.columns:
        conn.execute(f"CREATE INDEX IF NOT EXISTS idx_orders_{date_col} ON orders({date_col});")

print("SQLite DB created at:", db_path)
```

```python
q1 = """
SELECT
    COUNT(DISTINCT orderid) AS unique_orders,
    COUNT(*) AS lines_count,
    SUM(quantity) AS total_units,
    ROUND(SUM(sales), 2) AS total_revenue
FROM orders;
"""
pd.read_sql_query(q1, conn)
```

```python
q2 = f"""
SELECT
    strftime('%Y-%m', {date_col}) AS year_month,
    ROUND(SUM(sales), 2) AS revenue,
    SUM(quantity) AS units
FROM orders
GROUP BY 1
ORDER BY 1;
"""
trend_monthly = pd.read_sql_query(q2, conn)
trend_monthly.head(10)
```

```python
q5 = """
SELECT
    storeid,
    ROUND(SUM(sales), 2) AS revenue,
    SUM(quantity) AS units,
    COUNT(DISTINCT orderid) AS orders
FROM orders
GROUP BY storeid
ORDER BY revenue DESC;
"""
store_perf = pd.read_sql_query(q5, conn)
store_perf.head(10)
```

```python
q6 = """
WITH bucketed AS (
  SELECT
    CASE
      WHEN discount_pct IS NULL THEN '00 None'
      WHEN discount_pct < 5 THEN '01 [0-5)'
      WHEN discount_pct < 10 THEN '02 [5-10)'
      WHEN discount_pct < 20 THEN '03 [10-20)'
      WHEN discount_pct < 30 THEN '04 [20-30)'
      ELSE '05 30%+'
    END AS discount_bucket,
    sales,
    quantity
  FROM orders
)
SELECT
  discount_bucket,
  ROUND(SUM(sales), 2) AS revenue,
  SUM(quantity) AS units,
  COUNT(*) AS lines
FROM bucketed
GROUP BY discount_bucket
ORDER BY discount_bucket;
"""
disc_impact = pd.read_sql_query(q6, conn)
disc_impact
```

```python
q7 = f"""
WITH ranked AS (
  SELECT
    customerid,
    {date_col} AS order_date,
    orderid,
    ROW_NUMBER() OVER (PARTITION BY customerid ORDER BY {date_col}) AS rn
  FROM orders
)
SELECT
  SUM(CASE WHEN rn = 1 THEN 1 ELSE 0 END) AS first_orders,
  SUM(CASE WHEN rn > 1 THEN 1 ELSE 0 END) AS repeat_orders
FROM ranked;
"""
pd.read_sql_query(q7, conn)
```
